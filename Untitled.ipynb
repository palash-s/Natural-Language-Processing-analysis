```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/palas/Downloads/Customer_Reviews.csv")
data.head()
```

```python
words_per_rreview = data.Text.apply(lambda x: len(x.split(" ")))
words_per_rreview.hist(bins = 100)
plt.xlabel('Review words')
plt.ylabel('Frequency')
plt.show()
```

```python
print(words_per_rreview.mean())
print(words_per_rreview.skew())
```

```python
#distribution score
percent = 100* data['Score'].value_counts()/len(data)
```

```python
percent
```

```python
percent.plot.bar()
plt.show()
```

```python
from wordcloud import WordCloud
word_cloud = ''.join(data['Text'])

wc = WordCloud(max_font_size = 100, background_color = "white", scale = 10, width = 800, height = 400).generate(word_cloud)

plt.figure()
plt.imshow(wc, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
# mapping review

data['rating'] = np.where(data.Score > 3,1,0)

data = data[data.Score != 3]

data['rating'].value_counts()
```
